```python
import numpy as np
import pandas as pd
from 
```

```python
data = pd.read_csv("Sales.csv")
df = pd.DataFrame(data)
df = df.dropna()
df['Rating'] = df['Rating'].interpolate(method = 'linear')
print(df.describe())
df.info()
```

```python
X = df[['Brands' , 'Models' , 'Memory' , 'Camera' , 'Storage' , 'Discount' , 'Original Price']]
Y = df['Selling Price']
```

```python
#Encoding the categorical columns
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['Brands' , 'Models' , 'Memory' , 'Camera' , 'Storage']
numerical_features = ['Discount' , 'Orignal Price']

preprocessor = ColumnTransformer(
    transformers = [
        ('cat' , OneHotEncoder(handle_unknown = 'ignore') , categorical_features)
    ],
    remainder = 'passthrough'
)
X_encoded = preprocessor.fit_transform(X)
```

```python
#Train test split the data
# 80% Training 
# 20% Testing
from sklearn.model_selection import train_test_split
X_train , x_test , Y_train , y_test = train_test_split(
    X_encoded,
    Y,
    test_size = 0.2,
    random_state = 42
)
```

```python
#Now here we are going to train the model we Check for many Models like Regression , RandomForest , Decission Tress etc
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 200 , random_state = 42)

model.fit(X_train , Y_train)
```

```python
#checking the accuracy of our model
score = model.score(x_test , y_test)
print("Accuracy" , score)
```

```python
#Creating a function for price prediction
def predict_price(Brands ,Models , Memory ,Camera , Storage ,Discount , Original_Price):
    input_df = pd.DataFrame([[Brands ,Models , Memory ,Camera , Storage ,Discount , Original_Price]],
                           columns = ['Brands' , 'Models' , 'Memory' , 'Camera' , 'Storage' , 'Discount' , 'Original Price'])
    #encodied the inputs 
    encoded_inputs = preprocessor.transform(input_df)
    #prediction
    prediction = model.predict(encoded_inputs)
    return prediction[0]
    
    
```

```python
print(predict_price("Apple", "iPhone 11", "4GB", "64MP", "256GB", 0, 64900))
```
