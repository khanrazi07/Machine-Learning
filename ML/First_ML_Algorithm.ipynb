```python
#  We are making a model of machine learning to predict that the given fruit is an applle or not 

from sklearn.tree import DecisionTreeClassifier
#making a data set with features

features = [
    [150,0] , 
    [120,0],
    [60 ,1],
    [160 ,0],
    [180,1],
    [200,1],
    [250,0],
    [220,1]
]
labels = ["Apple" , "Apple" , "Orange" , "Apple" , "Orange" , "Orange" , "Apple" , "Orange"]

clf = DecisionTreeClassifier()
clf = clf.fit(features , labels)

for i in features:
    print(clf.predict([i]))
    
```
